```python
from bs4 import BeautifulSoup
import pandas as pd 
import requests
import numpy as np
```

```python
link='https://www.icc-cricket.com/rankings/womens/player-rankings/odi/all-rounder'
page=requests.get(link).text
```

```python
soup=BeautifulSoup(page,'html.parser')
```

```python
FirstPos=soup.find('span',class_='rankings-block__pos-number').text.split()
FirstAllroundName=soup.find('div',class_='rankings-block__banner--name-large').text
timNam=soup.find('div',class_='rankings-block__banner--nationality').text.split()
Rate=soup.find('div',class_='rankings-block__banner--rating').text
```

```python
TopAllRound_df=pd.DataFrame({'POSITION': FirstPos, 'BATSMAN NAME': FirstAllroundName, 'TEAM NAME': timNam,'RATING':Rate})
```

```python
AllRounder_woman_data=soup.findAll('tr',class_='table-body')
```

```python
position=[]
Batsman=[]
TeamName=[]
Rating=[]

for store in AllRounder_woman_data:
    bat=store.find('span',class_='rankings-table__pos-number').text.split()
    position.append(bat)
    
    BatName= store.find('td',class_='table-body__cell rankings-table__name name').text.replace('[' , '').split()
    Batsman.append(BatName)
    
    temName= store.find('span',class_='table-body__logo-text').text
    TeamName.append(temName)
    
    rat=store.find('td',class_='table-body__cell rating').text
    Rating.append(rat)
```

```python
TopAllRound_df2=pd.DataFrame({'POSITION': position, 'BATSMAN NAME': Batsman, 'TEAM NAME': TeamName,'RATING':Rating})
```

```python
topers=[TopAllRound_df,TopAllRound_df2]
TOP_WOMEN_ALL_ROUNDER=pd.concat(topers)
```
