```python
from bs4 import BeautifulSoup
import requests
```

```python
wiki_link="https://en.wikipedia.org/wiki/Main_Page"
page= requests.get(wiki_link).text
```

```python
soup=BeautifulSoup(page,'lxml')
print(soup.prettify())
```

```python
story = soup.find_all(['h1', 'h2','h3'])
```

```python
for header in story:
    print(header)
```
